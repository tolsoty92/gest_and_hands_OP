```python
columns = ['label']
for x in range(18):
    columns.append('kpoint x'+str(x))
    columns.append('kpoint y'+str(x))
print(columns)
```

```python
df_lst = []
df = pd.DataFrame(columns=columns)
for dir in os.listdir('./np_arrays'):
    df = pd.DataFrame(columns=columns)
    for array in os.listdir('./np_arrays/'+dir):
        f_name = './np_arrays/'+dir+'/'+array
        np_array = np.load(f_name)
        np_array = np.array([arr[:2] for arr in np_array])
        row = [dir]
        for kpoint in np_array:
            row.append(kpoint[0])
            row.append(kpoint[1])
        df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)
    df_lst.append(df)
```
